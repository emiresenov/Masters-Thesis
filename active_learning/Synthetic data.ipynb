```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

# Generate data

```python
xs = np.random.uniform(low=0.0, high=1000.0, size=2500)
ys = np.random.uniform(low=0.0, high=1000.0, size=2500)
```

## Create synthetic datasets

```python
x1 = np.arange(50, 1000, 0.1)
f1 = lambda x : 50000 / x
df1 = pd.DataFrame({'x' : xs, 'y' : ys})
df1['class'] = np.where(df1['y'] > f1(df1['x']), 1, 0)
```

```python
x = np.arange(0.1, 1000, 0.1)
f2 = lambda x : 250 * np.cos(0.03*x) + 500
df2 = pd.DataFrame({'x' : xs, 'y' : ys})
df2['class'] = np.where(df2['y'] > f2(df2['x']), 1, 0)
```

```python
x3 = np.arange(723, 777, 0.1)
f3 = lambda x : np.power(x - 750, 2) + 250
df3 = pd.DataFrame({'x' : xs, 'y' : ys})
df3['class'] = np.where(df3['y'] > f3(df3['x']), 1, 0)
```

```python
f4 = lambda x, y: np.power(x - 300, 2) + np.power(y - 300, 2) <= np.power(150, 2)
df4 = pd.DataFrame({'x' : xs, 'y' : ys})
df4['class'] = np.where(f4(df4['x'], df4['y']), 1, 0)
```

# Plot synthetic datasets

```python
plt.rcParams["figure.figsize"] = (10,7.5)
sns.set_theme()
```

```python
fig, axs = plt.subplots(2, 2)

# Plot synthetic dataset 1
sns.scatterplot(x='x', y='y', data=df1, hue='class', ax=axs[0,0]).set(xlabel=None, ylabel=None, title='Synthetic dataset 1')
axs[0,0].plot(x1, f1(x1), color='black')
axs[0,0].legend([],[], frameon=False)

# Plot synthetic dataset 2
sns.scatterplot(x='x', y='y', data=df2, hue='class', ax=axs[0,1]).set(xlabel=None, ylabel=None, title='Synthetic dataset 2')
axs[0,1].plot(x, f2(x), color='black')
axs[0,1].legend([],[], frameon=False)

# Plot synthetic dataset 3
sns.scatterplot(x='x', y='y', data=df3, hue='class', ax=axs[1,0]).set(xlabel=None, ylabel=None, title='Synthetic dataset 3')
axs[1,0].plot(x3, f3(x3), color='black')
axs[1,0].legend([],[], frameon=False)


# Plot synthetic dataset 4
sns.scatterplot(x='x', y='y', data=df4, hue='class', ax=axs[1,1]).set(xlabel=None, ylabel=None, title='Synthetic dataset 4')
circ = plt.Circle((300, 300), 150, color='black', fill=False, linewidth=1.5)
axs[1,1].add_patch(circ)
axs[1,1].legend([],[], frameon=False)

fig.tight_layout()

plt.show()
```
